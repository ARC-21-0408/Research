## 1.3 What is a parsec (pc)?  What is a Megaparsec (Mpc)?

A parsec is a unit of distance. A parsec can be calculated by measuring the paralax angle of the object of interest in arcseconds. The closer an object is to the Earth, the larger the paralax angle. Remember that an arcminute is 1/60th of a degree and an arcsecond is 1/60th of an arcminute. We want to convert the degrees to radians by simply multiplying the value in degrees by, π/180. One parsec is approximatly 3.08x10^16 m, 206264 AU, or 3.26 ly.  The distance in parsecs can be apporiximated by, 

D = R / θ

where D is the distance to the object in usually in AU (depends on what are the units of R), R is the distance from the Sun to the Earth (usually in AU's), and θ is the paralax angle in arcseconds. θ must first be converted to radians before making the calulation.

θ = (1/60)(x/60)(π/180)

where x is the value of the paralax angle. For example, if the angle was 2 arcseconds, (2"), then x=2. 
The distance between the Earth and the Sun is 1 AU, which makes the calculation of D much easier. Once D is calculated, it will have units of AU. To convert this value to parsecs we use the simple conversion factor 
1 {pc}/ 206264 {AU}, ({} is just denoting units) 

D {AU} * (1 {Pc} / 206264 {AU}) = D {Pc}

D is now in parsecs instead of AU.

Since stars are very very far away using km or AU is not an efficient way of measuring distances because they will be such large quantities. Instead of have to write 3.08x10^16 or some very large number, we can save lot of time by just writing 1 pc. 

A Mpc is simply one million (10^6) parsecs. The "M" stands for "mega" parsec, but its easier to remember that the "M" stands for "million".

## 1.4 What is the difference between weight and mass?

Mass is the amount of matter something contains. For example, the mass of a box is the amount of the matter of the box plus whatever is inside the box. Mass will NOT change with location. If you are on the moon, you have the same amount of mass than if you were standing on the Earth. 

Weight is the measure of the pull of gravity on the object. This is why you stand on a scale to get obtain your body weight. The scale measures how hard a body of mass is being pulled by gravity. Weight DOES change with location. You will weigh less on the moon because the force of gravity is smaller than the force of gravity on the Earth. 

Weight is determined by multiplying mass by the force of gravity. 

W = mg

where m is the mass of the object and g is the acceleration due to gravity. For the Earth g = 9.81 {m/s^2}. If you dont know the mass of an object, do not fear! Simply weigh the object and then divide the weight by g to get the total mass of the object. 

m = W/g



## 1.5 What is redshift?



The universe is expanding away from us in all directions, and the farther away an object such as a galaxy is from us, the faster it is moving away from us.  This expansion results in a change in the color of the distant object that we call the redshift (z).  Note that it is not the galaxy that changes color, but only the galaxy image.  This is analogous to putting on colored swimming goggles or sunglasses (blue).  When you wear the headgear your friends faces and skin may suddenly all look blue to you, but actually it is only your perception of their skin color that changed.  Redshift is always positive, such that higher values correspond to greater distance.  We measure the redshift by dispersing the light from a distant galaxy into its constituent rainbow of colors, and then by comparing that light signature with that of a nearby galaxy.  The greater the differences between the two galaxies, then the greater the redshift.

Knowing the redshift, we can estimate the distance to the source. The farther away the source the more distance the light has to travel so the greater the distance to greater redshift will be observed. 


Redshift can be approximated in terms of the velocity of the galaxy cluster, where c is the speed of light

z = v / c

Solving for v we an find a relationship to the Distance, D 

v = HoD

where Ho is hubble cosntant

Finally we can combine both equations to obtain an expression for distance to the galaxy cluster where:

D = zc/Ho



If you enter in a value for the redshift of galaxy cluster here: 

z = 0.1756

then we can solve for distance to galaxy cluster.

```python
#z = xxx        # enter your value for the redshift here.
Ho = 71         #Hubble's consant in km/s Mpc where Mpc is megaparsecs 
c = 299792458   #speed of light

D = (z*c)/(1000*Ho)  #calculating for Distance, must multiply Hubble's constant to convert km to m


print('Distance =', D, 'Mpc')
```

## 1.6 What is an angular radius?

Measurements of angles in space is very important as it is related to the physical size.  When we observe an object in space we can measure the radius θ in degrees (or equivalently how many arcseconds).  This is the angular radius. Knowing this, and the distance the object is away from us, we can also compute the actual radius. 
For smaller distances we can describe this relationship by a right triangle. By geometry, we can write:
tan(θ) = r/d
where θ is the angular radius, r is the actual radius and d is the distance to the object from the observer.
In astronomical applications we can further write that tan(θ) = θ, and we arrive at the final equation:

θ = r/d

This is pretty straightforward, right?  But if we measusure angular radius in arcseconds we have to convert to radians. To do this we can write:

θ = # of arcseconds * (1 radian/206265 arcseconds)

## 1.7 What is a galaxy cluster?

 A galaxy is a large collection of 10-100 billion stars whose motions we do not understand.  A galaxy cluster is a collection of up to hundreds galaxies whose motions we do not understand.  In a cluster, the galaxies orbit about the center much like bees circle a beehive. The bees circle at predictable speeds (but do not get too close)!  On the other had, the galaxies in a cluster orbit at speeds greater than expected, and in a way that can be explained if there is a large of amount of an unseen material that we call "dark matter."  This dark matter turns out to be a more substantial component of the cluster than that of all the stars in all of the member galaxies.  Wait - if dark matter is invisible, then how can we see it?  Well, just as wind is invisible but we know it is there when it shakes trees, dark matter is also invisible but know it is there because it speeds up the motions of the galaxies.  Trying to undertstand what is dark matter is one of the most important questions in all of astronomy and physics.

## 1.8 What is a gravitational lens, and a giant arc?

Einstein tells us that mass bends light.  A very massive object will change the path of the light passing by it, similar to how eyeglasses change the path of the light.  This "bending" of light is called gravitational lensing.  In this experiment we are concerned with natural "eyeglasses" in space called lenses.  These lenses take the shape of a galaxy cluster, so they are humongous eyeglasses!  These lenses are not polished like real eyeglasses, and so produce distorted images instead of in-focus ones, and they even make you "see double."  If this happened to a real pair of glasses then you would send them back to the shop.  How does a gravitational lens distort the light?  Well, the appearance of an otherwise circular-shaped object is bent into a banana-like shape that we call a gravitational arc, or just an "arc."  In exceptional circumstances, that banana-like shape is large relative to the other galaxies.  In such cases we call the image a "giant arc."  The aim of this experiment is to find the giant arcs, and to use them as a clue as to where and how big the natural lens is.

```python
pi = 3.14
Sigma_cr = 3.5e45 # kg/Mpc^2

#θ = Angular_size_radius_radians   # uncomment this line.
#D = xxx  #Insert value for distance to the galaxy cluster in Mpc here and then uncomment line.

#Now square D and θ:
Dsqu = 760*760  # enter code here.
θsqu = 4.8e-5*4.8e-5  # enter code here.

# The total mass, M is:

M = pi *Sigma_cr * Dsqu * θsqu  # in kg
#print("The mass =',M,'kg')
print(M)
```

To make more sense of this incredibly-large number, we usually quote our answers in terms of solar masses.  The mass of our Sun is 1.99 x 10$^{30}$ kg. which is is a factor of 300,000 times the mass of the Earth.  Can you verify this number for yourself?  

```python
Msun = 1.989*10**30 #Mass of Sun, in kg

Msolar = M/Msun # taking the mass of the galaxy cluster and converting it in terms of Solar mass

print('Mass =', Msolar, 'Solar Masses') 
```
