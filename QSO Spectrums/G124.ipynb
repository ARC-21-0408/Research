## Spectrum Analysis Code for QSO's

```python
import matplotlib
import matplotlib.pyplot as plt
fig = plt.figure()

import numpy as np  
from astropy.io import fits
from decimal import Decimal
matplotlib.rcParams.update({'font.size': 18}) 

#Read fits file
hdul = fits.open('spec_QSO1.fits', memmap=True)  
# hdul.info()
data = hdul[1].data
cols = hdul[1].columns
#cols.info() # prints out the names of the columns
#print(data[0]) # prints out the first row of data (first element of each of column)

# Extract the wavelength (in units of "log Angstroms"), and flux (in units of "10e-17 erg/s/cm^2/Angstrom") 
# Call the wavelength "wl" and the flux "flux".  
wavetmp = data.field('loglam')
wl = 10**wavetmp  
flux = data.field('flux') 
                     
# Plot up the QSO spectrum. 
# call the plot parameter of "drawstyle='steps-mid'" to see exactly where the data points are.

#G124
z1=2.1615 #cluster
z2=3.172016 #QSO1
#z2=2.268088 #QSO2
#z2=1.924046 #QSO3

#PLCK G218.9+55.5
z1= 2.43 #cluster
z2=1.991876 #QSO4


#Ly_alpha

x1=1216.67*(1+z1)

x2=1216*(1+z2)

#CIV
x3_1=1548.203*(1+z1)
x3_2=1550.777*(1+z1)

x4_1=1548.203*(1+z2)
x4_2=1550.777*(1+z2)

#MgII
x5_1=2796.352*(1+z1)
x5_2=2803.531*(1+z1)

x6_1=2796.352*(1+z2)
x6_2=2803.531*(1+z2)

fig, ax = plt.subplots(figsize=(16,8))
ax.plot(wl, flux,drawstyle='steps-mid')
plt.axvline(x1, color='red')
plt.text(x1,12,'LyA')
plt.axvline(x2, color ='green')
plt.text(x2,12,'LyA')
plt.axvline(x3_1, color='red')
plt.text(x3_1,12,'CIV_1')
plt.axvline(x3_2, color='red')
plt.text(x3_2,11,'CIV_2')
plt.axvline(x4_1, color ='green')
plt.text(x4_1,12,'CIV_1')
plt.axvline(x4_2, color ='green')
plt.text(x4_2,11,'CIV_2')
plt.axvline(x5_1, color='red')
plt.text(x5_1,11,'MgII_1')
plt.axvline(x5_2, color='red')
plt.text(x5_2,10,'MgII_2')
plt.axvline(x6_1, color ='green')
plt.text(x6_1,7,'MgII_1')
plt.axvline(x6_2, color ='green')
plt.text(x6_2,6,'MgII_2')
#plt.axhline(y=12,color='purple')
ax.set(xlabel='Observed Wavelength (Angstroms)', xscale = 'linear', ylabel='Flux (x 10^-17 erg/s/cm^2/Ang)', yscale = 'linear')
ax.set_xlim(3800, 9500) #full spec
#ax.set_xlim(3800, 4000) #Ly_alpha
#ax.set_xlim(4850, 5050) #CIV
#ax.set_xlim(8800, 9000) #MgII
ax.set_ylim(-2,13)

#fig.savefig('QSO4.png') 
#fig.savefig('QSO4_zoom_lyalpha.png')
#fig.savefig('QSO4_zoom_civ.png') 
#fig.savefig('QSO4_zoom_mgii.png')
```
