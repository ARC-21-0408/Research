```python
hdul = fits.open('spec_QSO1.fits', memmap=True)  
# hdul.info()
data = hdul[1].data
cols = hdul[1].columns
#cols.info() # prints out the names of the columns
#print(data[0]) # prints out the first row of data (first element of each of column)

# Extract the wavelength (in units of "log Angstroms"), and flux (in units of "10e-17 erg/s/cm^2/Angstrom") 
# Call the wavelength "wl" and the flux "flux".  
wavetmp = data.field('loglam')
wl = 10**wavetmp  
flux = data.field('flux') 
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(wl, flux,drawstyle='steps-mid')
plt.axvline(x1, color='red')
plt.text(x1,8,'LyA')

#plt.axvline(x2, color ='green')
#plt.text(x2,12,'LyA')
ax.set(xlabel='Observed Wavelength (Angstroms)', xscale = 'linear', ylabel='Flux (x 10^-17 erg/s/cm^2/Ang)', yscale = 'linear')
#fig.savefig('xyz.png')  #If you want to save the plot
ax.set_xlim(3800, 3900)
ax.set_ylim(-2,10)



wvlo=4380. #low end of peak#
wvmid=4438. #red line
wvhi=4515. #high end of peak#
Fc=2.0 #continuium level#
W=0.

def find_nearest(wv,value):
    idx= (np.abs(wv - value)).argmin()
    return idx

idxlo=(find_nearest(wl,wvlo))  #pixel at specfiic wavelength or flux value
idxmid=(find_nearest(wl,wvmid))
idxhi=(find_nearest(wl,wvhi))
idylo=(find_nearest(wl,wvlo))
idymid=(find_nearest(wl,wvmid))
idyhi=(find_nearest(wl,wvhi))

W=0.
W2=0.

R=wl[idxlo] #Starting wavelength [corresponding pixel at wavelength value]
R2=wl[idxmid] #End of Red section wavelength
Fl=flux[idylo] #starting flux value [corresponding pixel at flux value]
Fl2=flux[idymid] #end of red section flux value 
testlo=R
testhigh=R2

print( wl[idxlo], wl[idymid], wl[idxhi]) #check the start, middle, and end values#

#while loop that ends when R reaches end of red section#
while (R < wl[idxmid]):
    idtestlo=(find_nearest(wl,testlo)) #finds the nearest pixel at a given wavelegth
    Flo=flux[idtestlo] #value of flux at pixel found in idtestlo
    W += (Fc - Flo)/Fc #computes equivalent width
    testlo += 1.0 #increments wavelength by one and idtestlo will find pixel closest to new wavelength
    R += 1.0 #increases R and causes loop to terminate when R is greater than 4438
    #print(idtestlo, Flo , R) #use this to check if getting correct pixel and flux values
print(W) #print equivalent width#

```
